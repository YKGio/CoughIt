```python
import soundfile
import os
import random
import matplotlib.pyplot as plt
import librosa
```

```python
cough_paths = []
for root, dirs, files in os.walk('../audios/'):
    for f in files:
        cough_paths.append(os.path.join(root, f))

coughs = []  
for path in cough_paths:     
    cough, sr = librosa.load(path, sr=48000)
    coughs.append(cough)
    
cough1, _= soundfile.read('../audios/cough1-1.wav')
```

```python
random.choice(coughs)
```

```python
plt.plot(cough1)
```
